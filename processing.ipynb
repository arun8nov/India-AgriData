## EDA

### Top 7 RICE PRODUCTION State Data(Bar_plot)

```python
R_df = df[['State Name','RICE PRODUCTION (1000 tons)']].groupby('State Name').sum().sort_values(by='RICE PRODUCTION (1000 tons)',ascending=False).head(7).reset_index()
R_df
```

```python
plt.figure(figsize=(15,6))
sns.set_style('whitegrid')
colors = plt.cm.cividis(np.linspace(0, 1, len(R_df)))
bars = plt.bar(R_df['State Name'], R_df['RICE PRODUCTION (1000 tons)'],color = colors)
for i in range(len(bars)):
    plt.text(i,R_df['RICE PRODUCTION (1000 tons)'][i],R_df['RICE PRODUCTION (1000 tons)'][i], 
             ha = "center", fontweight = "bold", fontsize = 10)
plt.xlabel('State Name')
plt.ylabel('Rice Production (in Tonnes)')
plt.xticks(rotation=90)
plt.title('Top 7 RICE PRODUCTION STATES')
plt.show()
```
