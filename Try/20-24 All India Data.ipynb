```python
import pandas as pd
pd.set_option('display.max_rows', 103)
pd.set_option('display.max_columns', 50)
```

```python
df = pd.read_csv("All-India_-Crop-wise-Area,-Production-&-Yield.csv",index_col = 0)
df
```

```python
df = df[df.Season == "Total"]
df
```

```python
df.columns
```

```python
df_Area = df[['Area-2020-21', 'Area-2021-22', 'Area-2022-23',
                'Area-2023-24', 'Area-2024-25']]

df_Area.columns = ['2020', '2021', '2022', '2023',
                    '2024']
df_Area = df_Area.T

df_Area.index.name = 'Year'

df_Area.columns = [i.upper() + " " +"AREA" for i in df_Area.columns]

df_Area
```

```python
df_Production = df[['Production-2020-21',
       'Production-2021-22', 'Production-2022-23', 'Production-2023-24',
       'Production-2024-25']]

df_Production.columns = ['2020', '2021', '2022', '2023',
                    '2024']
df_Production = df_Production.T

df_Production.index.name = 'Year'

df_Production.columns = [i.upper() + " " +"PRODUCTION" for i in df_Production.columns]

df_Production
```

```python
df_Yield = df[['Yield-2020-21', 'Yield-2021-22', 'Yield-2022-23',
       'Yield-2023-24', 'Yield-2024-25']]

df_Yield.columns = ['2020', '2021', '2022', '2023',
                    '2024']
df_Yield = df_Yield.T

df_Yield.index.name = 'Year'

df_Yield.columns = [i.upper() + " " +"YIELD" for i in df_Yield.columns]

df_Yield
```

```python
l = []
i=0
while i < 37:
    l.append(df_Area.columns[i])
    l.append(df_Production.columns[i])
    l.append(df_Yield.columns[i])
    i = i+1
print(l)
```

```python
df1 = df_Area.join(df_Production).join(df_Yield)
df1[l]
```

```python
df1.to_csv("All-India-20 to 24 Data.csv")
```
